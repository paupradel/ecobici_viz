# Estaciones de ecobici con ageb, colonia y código postal

```python
import plotly.graph_objects as go
import pandas as pd
import geopandas as gpd
import json
```

```python
df_sankey= pd.read_csv('./data/production_data/viajes_ecobici_entre_estaciones.csv')
```

```python
df_1 = df_sankey[['CVE_AGEB_arribo', 'nombre_estacion_arribo']]
```

```python
df_1.sort_values(by=['CVE_AGEB_arribo'], inplace=True)
```

```python
df_1.reset_index(inplace=True, drop=True)
```

```python
df_1.drop_duplicates(inplace=True)
```

```python
df_1.reset_index(inplace=True, drop=True)
```

```python
df_1 = df_1.rename(columns={"CVE_AGEB_arribo": "CVE_AGEB", "nombre_estacion_arribo": "text"})
```

```python
df_1.head()
```

```python
df_1.to_csv('estaciones-agebs.csv', index=False)
```

```python
df_1 = pd.read_csv('estaciones_en_agebs.csv')
```

```python
df_1.head()
```

```python
df_1 = df_1.loc[df_1['CVE_AGEB'].isin([5,37])]
```

```python
estaciones_primer_ageb = df_1.iloc[0]['text']
```

```python
estaciones_segundo_ageb = df_1.iloc[1]['text']
```

```python
df_1['text'] = df_1[['CVE_AGEB','text']].groupby(['CVE_AGEB'])['text'].transform(lambda x: ','.join(x))
```

```python
df_agebs = df_1[['CVE_AGEB', 'text']].drop_duplicates()
```

```python
df_agebs.reset_index(inplace=True, drop=True)
```

```python
df_agebs.to_csv('estaciones_en_agebs.csv', index=False)
```
